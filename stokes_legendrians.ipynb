```python
import matplotlib.pyplot as plt
import numpy as np
from sage.all import *
%matplotlib widget
```

```python
r = 50000000


x = var('x', domain=CC)
t = var('t', domain=CC)

# exponents = 2/3*r*(3/2)*np.exp(1j*3/2*x)

def stokes_link(expression, num=800):
    theta =  expression.variables()[0]
    domain = np.linspace(0, 2*np.pi, num=num)
    values = np.array([expression.subs(theta==val).n() for val in domain])

    positive = np.real(values)
    negative = np.real(-values)

    plt.close()
    plt.plot(domain, positive)
    plt.plot(domain, negative)
    plt.grid(True)
    plt.show()

def expo(coefficient, angle_variable=x, radius=r):
    return radius^(coefficient)*np.exp(1j*coefficient*angle_variable)

def stokes_link_from_exponent(expression, num=800):
        stokes_link(expression=expression.integrate(x), num=num)
```

# Airy Equation

```python
integral_airy = 2/3*expo((3/2))
exponents_airy = expo(1/2)
stokes_link(integral_airy)
stokes_link_from_exponent(exponents_airy)
```

# PI

```python
# exponents_p1 = exp0 + t/2*r*np.exp(1j*x*1/2)
exponents_p1 = expo(5/2) + t/2*expo(1/2)

stokes_link(exponents_p1.subs(t==100))
```

# PII

```python
theta_infinity = var('theta_infinity', domain=CC)
exponents_p2 = expo(3) + t/2 * expo(1) + theta_infinity/2

stokes_link(exponents_p2.subs({t:1, theta_infinity:2.5}))
```
